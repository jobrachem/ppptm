```python
import ppptm as gptm
```

```python
data = gptm.load_americas()
```

```python
locs = gptm.LocationVars.new_from(data.locs, n_subset=512)
model = gptm.Model.new_G(data.obs, locs)
```

```python
obs_mean = data.obs.mean(axis=0, keepdims=True)
df = gptm.long_df(
    locs.locs.unordered,
    obs_mean,
)
gptm.plot_df(df)
```
